```r
# annotaion for color
data_anno <- readRDS("/faststorage/project/jsp_student_projects/shared_data/TCGA_counts/TCGA_sample_anno_match.rds")

# Fold change:
df <- readRDS("~/jsp_student_projects/shared_data/TCGA_counts/TCGA_isoform_counts/TCGA_mirna_TPMs_iso_match.rds")
mat <- as.matrix(t(df))
head(df)
```

```r
PCA <- RunPCA(mat, assay = NULL, npcs = 50, rev.pca = F, weight.by.var = T, verbose = F)
c_cells_tsne <- RunTSNE(
  PCA[],
 # mat, # for representations
  assay = "RNA",
  seed.use = 1,
  tsne.method = "Rtsne",
  dim.embed = 2,
  max_iter = 500)
```

```r
tsne <- as.data.frame(c_cells_tsne@cell.embeddings)

tSNE_1_mRNA<- tsne[1]
tSNE_2_mRNA<- tsne[2]
to_plot <- cbind( tSNE_1_mRNA,tSNE_2_mRNA)

to_plot <- to_plot %>%
rename(tSNE_1_mRNA = tSNE_1,
        tSNE_2_mRNA = tSNE_2 )
head(to_plot)
```

```r
sample_id <- colnames(df)
sample_row <- as.data.frame(sample_id)
head(sample_row)
combined <- cbind(sample_row,tsne)
combined <- combined %>%
  left_join(data_anno, by = "sample_id")
head(combined)
```

```r
# Plot
options(repr.plot.width = 12, repr.plot.height = 8) 
combined %>%
  ggplot(aes(x = tSNE_1, y = tSNE_2, color = as.factor(max_prob_density_comp))) +
  geom_point() +
  scale_alpha_manual(values = c(0.05, 1)) +
  labs(title = "Latent representations visualized with tSNE ") + 
  labs(color = "Component no.", alpha = "Metastatic")+ 
  theme(plot.title = element_text(size=25))

#saveRDS(combined,"/faststorage/project/jsp_student_projects/miRNA_DGD_DE_F2024/processed_data/mRna_tSNE.rds" )
#ggsave(filename = "fold_change_scores_tsne2_tissue_type.png", width = 16, height = 12, path = "./plots")
```
